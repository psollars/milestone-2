```python
query = f"SELECT * FROM '{SCRATCH_DIR}/bts_faa_coords.parquet' ORDER BY RANDOM() LIMIT 100;"

bts_faa_coords_sample = duckdb.query(query).df()

bts_faa_coords_sample.to_csv("bts_faa_coords_sample.csv")

bts_faa_coords_sample.head()
```

```python
query = f"SELECT * FROM '{SCRATCH_DIR}/all_features_2019.parquet' LIMIT 100000;"

all_features_2019_df = duckdb.query(query).df().drop("__index_level_0__", axis="columns")

all_features_2019_sample = all_features_2019_df.sample(100, random_state=42)

all_features_2019_sample.to_csv("all_features_2019_sample.csv")

all_features_2019_sample.head()
```
